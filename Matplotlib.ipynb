```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import numpy as np

x = np.linspace(0, 10, 100)
x
```

```python
siny = np.sin(x)
cosy = np.cos(x)
```

```python
plt.plot(x, siny)
plt.plot(x, cosy)
plt.show()
```

```python
plt.plot(x, siny, color='red')
```

```python
plt.plot(x, siny, linestyle=':') #linestyle: :--.--
plt.plot(x, cosy)
plt.show()
```

```python
plt.plot(x, siny, color='red', linestyle=':') #linestyle: :--.--
plt.plot(x, cosy)
plt.xlim(-5, 15) #x
plt.ylim(0, 1.5)
plt.show()
```

```python
plt.plot(x, siny, color='red', linestyle=':') #linestyle: :--.--
plt.plot(x, cosy)
plt.axis([-1, 11, -2, 2]) #x,y
plt.show()
```

```python
plt.plot(x, siny, color='red', linestyle=':', label='sin(x)') #linestyle: :--.--
plt.plot(x, cosy, label='cos(x)')
plt.xlabel('x axis') #x
plt.ylabel('y axis')
plt.legend() #
plt.show()
```

```python
plt.plot(x, siny, color='red', linestyle=':', label='sin(x)') #linestyle: :--.--
plt.plot(x, cosy, label='cos(x)')
plt.xlabel('x axis') #x
plt.ylabel('y axis')
plt.legend() #
plt.title('Machine Learning')
plt.show()
```
