```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import numpy as np

x = np.linspace(0, 10, 100)
x
```

```python
siny = np.sin(x)
cosy = np.cos(x)
```

```python
plt.plot(x, siny)
plt.plot(x, cosy)
plt.show()
```

```python
plt.plot(x, siny, color='red')
```

```python
plt.plot(x, siny, linestyle=':') #linestyle四种样式: :点虚线、-、-.、--
plt.plot(x, cosy)
plt.show()
```

```python
plt.plot(x, siny, color='red', linestyle=':') #linestyle四种样式: :点虚线、-、-.、--
plt.plot(x, cosy)
plt.xlim(-5, 15) #x轴取值范围
plt.ylim(0, 1.5)
plt.show()
```

```python
plt.plot(x, siny, color='red', linestyle=':') #linestyle四种样式: :点虚线、-、-.、--
plt.plot(x, cosy)
plt.axis([-1, 11, -2, 2]) #同时调节x,y轴坐标
plt.show()
```

```python
plt.plot(x, siny, color='red', linestyle=':', label='sin(x)') #linestyle四种样式: :点虚线、-、-.、--
plt.plot(x, cosy, label='cos(x)')
plt.xlabel('x axis') #x轴标签
plt.ylabel('y axis')
plt.legend() #添加图示
plt.show()
```

```python
plt.plot(x, siny, color='red', linestyle=':', label='sin(x)') #linestyle四种样式: :点虚线、-、-.、--
plt.plot(x, cosy, label='cos(x)')
plt.xlabel('x axis') #x轴标签
plt.ylabel('y axis')
plt.legend() #添加图示
plt.title('Machine Learning')
plt.show()
```

### Scatter Plot 对于折线图一般横轴是特征 对于散点图则是横纵两个轴都是特征值

```python
plt.scatter(x, siny)
plt.show()
```

```python
plt.scatter(x, siny)
plt.scatter(x, cosy, color='red')
plt.show()
```

```python
x = np.random.normal(0, 1, 10000)
y = np.random.normal(0, 1, 10000)

plt.scatter(x, y,alpha=0.5) #alpha代表透明度，0表示完全透明 1表示完全不透明
plt.show()
```
