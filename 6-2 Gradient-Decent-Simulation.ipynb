```python
import numpy as  np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1, 6, 141)
x
```

```python
y = (x - 2.5) **2 - 1
y
```

```python
plt.plot(x, y)
plt.show()
```

```python
def J(theta):
    return (theta - 2.5) **2 - 1
def dJ(theta):
    return 2 * theta - 5
```

```python
eta = 0.1
theta = 0.0
epsilon = 1e-8

theta_history = []

while True:
    
    gradient = dJ(theta)
    last_theta = theta
    theta_history.append(theta)
    theta = theta - eta * gradient
    
    if(abs(J(theta) - J(last_theta)) < epsilon):
        break
```

```python
len(theta_history)
```

```python
plt.plot(x, y, color='b')
plt.plot(theta_history, J(np.array(theta_history)), color='r', marker='+')
plt.show()
```

```python
def gradient_decent(initial_theta, eta, n_iters=1e4, epsilon=1e-8):
    
    
    theta = initial_theta
    theta_history.append(theta)
    i_iter = 0
    
    while i_iter < n_iters:
        gradient = dJ(theta)
        last_theta = theta
        theta_history.append(theta)
        theta -= eta * gradient
        
        if(abs(J(theta) - J(last_theta)) < epsilon):
            break
            
        i_iter += 1
```
